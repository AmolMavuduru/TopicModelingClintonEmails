```python
import re # regex library
def preprocessor(text):
    text = re.sub('<[^>]*>', '', text) # Effectively removes HTML markup tags
    emoticons = re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)', text)
    text = re.sub('[\W]+', ' ', text.lower()) + ' '.join(emoticons).replace('-', '')
    return text
```

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
emails['Processed Text'] = emails['RawText'].apply(preprocessor)
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
tfidf = TfidfVectorizer(strip_accents=None, lowercase=False, ngram_range=(1,1))
emails['vectors'] = tfidf.fit_transform(emails['Processed Text'])
```

```python
list(tfidf.transform([emails['Processed Text'].iloc[0]]).todense().flat)
```

```python
tfidf_matrix = []
for email in emails['Processed Text']:
    tfidf_output = list(tfidf.transform([email]).todense().flat)
    tfidf_matrix.append(tfidf_output)
```

```python
import time
start = time.time()
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(np.array(tfidf_matrix))
end = time.time()
print('Time elapsed: {}'.format(end - start))
```

```python
def get_frequency_in_hillary_emails(word):
    
    freq = 0
    for text in emails['Processed Text']:
    
        freq_text = text.count(word)
        freq += freq_text
    return freq
```

```python
get_frequency_in_hillary_emails('russia')
```

```python
freq_dict = []
i = 0
for word in tfidf.vocabulary_.keys():
    if i % 1000 == 0:
        print('Working on word {}'.format(i))
    freq_dict.append((word, get_frequency_in_hillary_emails(word)))
    i += 1
freq_dict = dict(freq_dict)
```

```python
from operator import itemgetter
from collections import OrderedDict

sorted_word_freqs = OrderedDict(sorted(freq_dict.items(), key=itemgetter(1), reverse=True))
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
STOP_WORDS
```

```python
top_hundred_words = []
i = 0
while(len(top_hundred_words) < 100):
    word, freq = list(sorted_word_freqs.items())[i]
    if word not in STOP_WORDS and len(word) >= 6:
        top_hundred_words.append((word, freq))
    i += 1
```
